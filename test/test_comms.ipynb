```python
import sys

sys.path.append("/home/lokgar/projects/schemdraw")

import schemdraw
from schemdraw import comms

schemdraw.use("svg")
```

```python
def drawElements(elm_list, n=5, dx=1, dy=2, ofst=0.8, fname=None, **kwargs):
    x, y = 0, 0
    d = schemdraw.Drawing(fontsize=12, font="Arial")
    for e in elm_list:
        element = getattr(comms, e)
        d += element().at(((d.unit + 1) * x + 1, y)).label(e, loc="top", ofst=0.1)
        x = x + dx
        if x >= n:
            x = 0
            y = y - dy
    return d


elems = [
    "Rectangle",
    "Fiber",
    "PolCtrl",
    "VOA",
    "Mod",
    "OSA",
    "ESA",
    "AWG",
    "Scope",
    "OPM",
    "PD",
    "LD",
    "Circle",
    "Square",
    # "Arrow",
    # "Line",
    "Filter",
    "Amp",
    "Circulator",
    "Adc",
    "Dac",
    # "Multiplexer",
    # "Arc2",
    # "Arc3",
    # # "ArcLoop",
    # "ArcN",
    # "ArcZ",
]
```

```python
display(drawElements(elems, d="right"))
```
